#### Import all the libraries that are necessary.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Define your own data. We would like to have a variable callex which contains a list of numbers from 0 to 100. Furthermore, we would like a variables called y which is twice x, and a variable called z which is the square of x.

```python
x = [x for x in range(0,101)]
y = [y*2 for y in x]
z = [z**2 for z in x]
```

#### Plot (x,y) and (x,z) on the axes in a line plot.

#### There are 2 ways of doing this - the way we have seen in class, and also using a subplot. Please research and try both!

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
plt.plot(x,y,color="#C29D4E",linewidth=4,linestyle="dashed")
plt.plot(x,z,color="#EF6555",linewidth=4,linestyle="dashed")
ax = plt.gca()
ax.set_facecolor('#253E66')
plt.show()
```

**subplotting**

```python
plt.style.use("default")
plt.subplot(1,2,1) #rows,columns,position
plt.plot(x,y,color="#C29D4E",linewidth=4,linestyle="dashed")
ax = plt.gca()
ax.set_facecolor('#253E66')
plt.title("X and Y")

plt.subplot(1,2,2) #rows,columns,position
plt.plot(x,z,color="#EF6555",linewidth=4,linestyle="dashed")
ax = plt.gca()
ax.set_facecolor('#253E66')
plt.title("X and Z")


plt.tight_layout()
plt.show()
```

#### Augmenting your previous code, label your axes.

```python
plt.style.use('default')
plt.subplot(1,2,1)
plt.plot(x,y,color="#C29D4E",linewidth=4,linestyle="dashed")
ax = plt.gca()
ax.set_facecolor('#253E66')
plt.title("X and Y")
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.subplot(1,2,2)
plt.plot(x,z,color="#EF6555",linewidth=4,linestyle="dashed")
ax = plt.gca()
ax.set_facecolor('#253E66')
plt.title("X and Z")


plt.tight_layout()
ax.set_xlabel('X')
ax.set_ylabel('Z')
plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.plot(x, np.exp(x))
plt.title("Normal scale")

plt.subplot(1, 2, 2)
plt.plot(x, x**2)
plt.plot(x, np.exp(x))
plt.yscale("log")
plt.title("Logarithmic scale (y)")

plt.subplots_adjust(right = 1.5)
plt.show()
```

```python

fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2,x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2,x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)")

plt.show()
```

### Using DataFrames

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
vehicles = pd.read_csv("vehicles.csv")
```

```python
grouped = vehicles[["Year","Combined MPG","Highway MPG"]].groupby("Year").sum().reset_index()
grouped.head()
```

```python
plt.style.use("default")
grouped.plot.scatter(x="Combined MPG",y="Highway MPG",c="#EF6555")
ax = plt.gca()
ax.set_facecolor('#253E66')
plt.tight_layout()
plt.show()
```

```python
vehicles.plot.scatter(x = "Combined MPG", y = "Highway MPG", c="Cylinders")
plt.show()
```

```python
vehicles.plot.scatter(x = "Combined MPG", y = "Highway MPG")
plt.show()
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
plt.style.use("ggplot")
grouped1 = vehicles[["Fuel Type","City MPG"]].groupby(["Fuel Type"]).mean().reset_index()
grouped1.plot.bar(x="Fuel Type",y="City MPG")
plt.xlabel("Fuel Type")
plt.title("City MPG by Fuel Type")
plt.tight_layout()
plt.show()
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
plt.hist(vehicles["Fuel Barrels/Year"],bins=20)
plt.show()
```
