```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vehicles = pd.read_csv('../../03_data-visualization/02_lab-matplotlib-seaborn/your-code/vehicles.csv')
vehicles.head()
```

```python
plt.scatter(vehicles["City MPG"],vehicles["Combined MPG"])
```

**Linear Correlation between two variables**

```python
vehicles["City MPG"].corr(vehicles["Combined MPG"]) # high correlation
```

**Correlation Matrix**

```python
vehicles.corr()
```

**Heatmap Correlation**

```python
sns.heatmap(vehicles.corr())
```

```python
sns.heatmap(vehicles.corr(), vmin=-1, vmax=1, annot=True, cmap="YlGnBu")
```

```python
vehicles.corr(method="spearman")
```

```python
sns.heatmap(vehicles.corr(method="spearman"), vmin=-1, vmax=1, annot=True, cmap="YlGnBu") 
#sorting the df and  the variables and
```

# Pivot Table

```python
help(pd.pivot_table)
```

```python
vehicles.pivot_table(index="Vehicle Class") # mean is the default
```

```python
vehicles.pivot_table(index="Drivetrain")
```

```python
vehicles.pivot_table(index="Drivetrain",aggfunc=np.median) #np.min,np.max
```

```python
vehicles.pivot_table(index="Drivetrain",aggfunc=len) #count
```

```python
vehicles.pivot_table(index="Drivetrain",aggfunc=len,columns="Year") #her yil icin count
```

```python
vehicles.pivot_table(index="Drivetrain",aggfunc=len,columns="Year",fill_value=0) # fill_value, NaN 0 oluyor
```

```python
vehicles.pivot_table(index="Drivetrain",aggfunc=len,columns="Year",values="City MPG",fill_value=0) # sadece City MPG yillara ve drive traine göre
```

```python
vehicles.pivot_table(index=["Vehicle Class","Drivetrain"],values="City MPG",fill_value=0) # 2index
```
