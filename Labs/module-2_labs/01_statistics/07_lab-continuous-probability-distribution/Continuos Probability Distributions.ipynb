```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Continuous Uniform Distribution**

```python
cuni = scipy.stats.uniform()
```

```python
cuni.rvs(size=10) #sample of 10 values between 0-1
```

```python
X = cuni.rvs(1000)
plt.hist(X,bins=100) #her deger arasinda 100 bin var 0 ile 0.2 arasinda
plt.show()
```

```python
#help(np.linspace)
```

```python
np.linspace(-1,1,1000) #-1 ile 1 arasinda 1000 tane deger verir
```

```python
numrange = np.linspace(-0.1,1.1,100) #pmf gibi
plt.plot(numrange,cuni.pdf(numrange)) #probability density function #what is the change the number in numrange is between 0 and 1
```

```python
plt.plot(numrange,cuni.cdf(numrange)) # cumulative density function #cmf gibi
```

```python
cuni = scipy.stats.uniform(0,10)
numrange = np.linspace(-1,11,100)
plt.plot(numrange,cuni.pdf(numrange),"o")
```

**Normal Distribution**

```python
from scipy.stats import norm
```

```python
normal = norm(0,1) #mean:0, standart_deviation:1
```

```python
numrange = np.linspace(-5,5,100)
plt.plot(numrange,normal.pdf(numrange))
```

```python
plt.plot(numrange,normal.cdf(numrange))
```

```python
X = normal.rvs(100000)
plt.hist(X,bins=1000)
plt.show()
```
