```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame()
df['x'] = np.arange(0,10,1)
df['y'] = df['x']**2
```

```python
df.plot("x","y")
plt.show()
```

```python
sns.lineplot("x","y",data=df)
plt.show()
```

```python
sns.lmplot("x","y",data=df)
#for regressions
#fit_reg=False, regression line i kaldiriyor
plt.show()
```

```python
data = pd.read_csv("./02_lab-matplotlib-seaborn/your-code/vehicles.csv")
```

```python
data.head()
```

**Boxplot**

**matplotlib**

```python
data.boxplot(column= "Combined MPG")
plt.show()
#there are outliers on the higher end of the distribution
#so the mean is higher than the median (income distribution)
```

```python
data["Combined MPG"].mean()
```

```python
data["Combined MPG"].median()
```

**In Seaborn**

```python
sns.boxplot("Combined MPG",data=data)
plt.show()
```

**Boxplot of 2 variables**

```python
sns.boxplot(x="Combined MPG",y="Highway MPG",data=data)
plt.show()
```

**Violinplot**

```python
sns.violinplot("Combined MPG",data=data)
plt.show()
```

```python
sns.violinplot(x="Combined MPG",y="Highway MPG",data=data)
plt.show()
```

**Histogram**

```python
plt.hist(data["Combined MPG"],bins=50)
plt.show()
```

```python
sns.distplot(data["Combined MPG"],bins=50)
plt.show()
```

**Bar Plot**

```python
sns.barplot("Highway MPG","Fuel Type",data=data) #default is the mean
plt.show()
```

```python
sns.barplot("Highway MPG","Fuel Type",data=data,estimator=np.median) #for median, for sum np.sum
plt.show()
```

```python
#plt.savefig() before the plt.show()
```

```python
fig = sns.barplot("Highway MPG","Fuel Type",data=data,estimator=np.median,saturation=0.9).get_figure()
fig.savefig("example.png")
```
